# visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.pairplot(data,hue='diagnosis',palette='Set1')## how each columns are related with each other
plt.show()
```

```python
print(data['diagnosis'].value_counts())##  This line counts the number of occurrences of each unique value in the diagnosis column of the data DataFrame.
sns.countplot(x='diagnosis',data=data)
plt.title('Diagnosis Distribution')
plt.xlabel("diagnosis")
plt.ylabel('Count')
plt.show()
```

```python
# conclusion
#The analysis of the diagnosis distribution reveals that there is a significantly higher count of samples labeled as 0 (benign) 
#compared to those labeled as 1 (malignant).  that means
#there are significantly more people who do not have cancer (represented by the label 0 for benign cases) 
#compared to those who do have cancer (represented by the label 1 for malignant cases).
```

```python
diagnosis_counts = data['diagnosis'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie(diagnosis_counts, labels=['Benign (0)', 'Malignant (1)'], autopct='%1.1f%%', startangle=90)
plt.title('Diagnosis Distribution')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular
plt.show()
```

```python
#conclusion
#The pie chart shows that there is a higher percentage of benign cases (labeled as 0) 
#compared to malignant cases (labeled as 1). Specifically, 62.7% of the samples are benign, while only 37% are malignant. This indicates that most individuals in the dataset do not have cancer.
```

```python
X=data.drop(columns=['diagnosis'])
Y=data['diagnosis']
```

# split the data into traning and test data set

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
Rf=RandomForestClassifier();
Rf.fit(x_train, y_train)
```

```python
predict=Rf.predict(x_test)
```

```python
predict
```

```python
accuracy_score(predict,y_test)
```
